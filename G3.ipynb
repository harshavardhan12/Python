```
def printeven(n):
 
  cnt=0;
  sum=0;
  while(cnt!=n):
    if(cnt%2==0):
      sum=sum+cnt;
    cnt=cnt+1;
  return sum;  
print(printeven(20));    
```

# Example
Factors

```
def factorslist(n):
  i=1;
  while(i!=n):
    if(n%i==0):
      print(i,end=" ");
    i=i+1;
  return;
factorslist(58);
```
