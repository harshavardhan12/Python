#Example 1
Function Programming to print the n Natural numbers

```
def printNNaturalNumbers(n):
  cnt=1;
  while(cnt<=n):
    print(cnt,end=" ");
    cnt=cnt+1;
  print()
  return

printNNaturalNumbers(9);
printNNaturalNumbers(15);
```

#Example 2
Find the factorial of given number


```
def findFact(n):
  fact = 1;
  while(n!=0):
    fact=fact*n;
    n=n-1;
  return fact;

print(findFact(5));
print(findFact(10));
print(findFact(20));
```

# Example 3
Program to find the count of palindromes between two limits 
## Test cases 
                         10,30     - - - 2

```
def countOfPalindromes(n1,n2):
    c=0;
    if(n1<n2):
        up=n2;
        low=n1;
    while(low!=up):
        i=low;
        r=0;
        while(i!=0):
            s=i%10;
            r=(r*10)+s;
            i=i//10;
        if(r==low):
            c=c+1;
        low=low+1;
    return c;
print(countOfPalindromes(1,100));
```
